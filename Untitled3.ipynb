PROPERTY COLUMN :properties_x and properties_y HAVE KEY VALUE PAIRS. WE EXTRACT THE PROPERTIES FROM THE COLUMNS AND JOIN EACH PROPERTY COLUMN TO MERGED_DF DATAFRAME.

```python

# Extract the properties from 'properties_x' and 'properties_y' columns
properties_df = pd.concat([merged_df['properties_x'].apply(eval).apply(pd.Series), merged_df['properties_y'].apply(eval).apply(pd.Series)], axis=1)

# Merge the extracted properties DataFrame with the original DataFrame
merged_df = pd.concat([merged_df, properties_df], axis=1)

# Drop the 'properties' and 'properties_y' columns from the DataFrame
merged_df.drop(['properties_x', 'properties_y'], axis=1, inplace=True)
```

WE NEED TO ADD A COLUMN  THE "FINAL RESULT" THAT IS,TO CHECK IF THE COSTUMER HAS PASSED OR FAILED. IN CASE HE IS PASSED IS IT IN HIS FIRST ATTEMPT OR SECOND ATTEMPT.By adding the "Final Result" column, you can easily identify whether a customer has passed or failed the KYC process and determine whether they passed on their first or second attempt. This additional information can provide valuable insights for further analysis and decision-making.

```python

def get_final_result(row):
    if row['attempt_id_x'] == row['attempt_id_y'] and row['result'] == 'clear':
        return 'first_attempt'
    elif row['attempt_id_x'] != row['attempt_id_y'] and row['result'] == 'clear':
        return 'second_attempt'
    else:
        return 'failed'

merged_df['final_result'] = merged_df.apply(get_final_result, axis=1)

```
